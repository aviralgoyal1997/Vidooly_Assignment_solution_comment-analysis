```python
import json
import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
from nltk.stem.snowball import SnowballStemmer
sbEng = SnowballStemmer('english')
```

```python
import re
with open("subtitles.txt", "r") as f, open("new_file.txt", "w") as n:
    x = f.read()
    result = re.sub("[^a-z\s]", "", x, 0, re.IGNORECASE | re.MULTILINE)
    n.write(result)
a=[]
with open("new_file.txt") as file:
    content = [x.strip('\n') for x in file.readlines()]
    content=' '.join(content)
    subtitles=content.split('  ')
        
```

```python
subtitle=[]
for i in subtitles:
    if(i!=''):
        subtitle.append(i)
```

Now we have subtitles in proper form and we will do preprocessing part

```python
def clean(sen):#to clean sentence
    sen=' '.join([item.strip('!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~') for item in (sen).lower().split(' ')]) #removing punctuation marks
    sen=' '.join([item for item in (sen).split(' ') if item not in stop])#stopwords removed
    sen=' '.join([sbEng.stem(item) for item in (sen).split(' ')]) #stemming,can slso perform lemmatizing
    return sen
```

```python
clean_subtitle=[]
for i in subtitle:
    clean_subtitle.append(clean(i))
```

```python
clean_subtitle #subtitles after preprocessing
```

Now we need to perform same things with comments and later will convert both to embeddings.

```python
emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

with open("comments.txt", "r") as f:
    x=f.readlines()
comments=[]        
for i in x:
    c=json.loads(i)
    comments.append(emoji_pattern.sub(r'',c['text'] ))
print (comments[0:5])
    
```

```python
clean_comment=[]
for i in comments:
    clean_comment.append(clean(i))
```

```python
def build_avg_vec(sentence, num_features):
    featureVec = np.zeros((num_features,), dtype="float32")
    nwords = 0
    
    for word in sentence.split():
        nwords += 1
        if word in vocab:
            featureVec = np.add(featureVec, model[word])
    if(nwords>0):
        featureVec = np.divide(featureVec, nwords)
    return featureVec
```

```python
import numpy as np
import pickle
with open('word2vec_model', 'rb') as handle:
    model = pickle.load(handle)
# vocab = stores all the words that are there in google w2v model
vocab = model.keys()
w2v_subtitle_weight = []
# for every subtitle build a weighted vector representation
for i in clean_subtitle:
    w2v_subtitle_weight.append(build_avg_vec(i, 300))
w2v_subtitle_weight = np.array(w2v_subtitle_weight)
# for every subtitle build a weighted vector representation
```

w2v_subtitle_weight contains subtitles embeddings


```python
w2v_comment_weight = []
# for every comment build a weighted vector representation
for i in clean_comment:
    w2v_comment_weight.append(build_avg_vec(i, 300))
w2v_comment_weight = np.array(w2v_comment_weight)
```

now to find similarity use euclidean dstance measure.

```python
matrix=np.zeros((len(clean_subtitle),len(clean_comment)))
```

```python
w2v_comment_weight[5]
```

```python
for i in range(len(w2v_subtitle_weight)):
    for j in range(len(w2v_comment_weight)):
        matrix[i][j]=np.linalg.norm(w2v_subtitle_weight[i]-w2v_comment_weight[j]) #computing euclidean distance
        
```

Now matrix have values representing similarity between each subtitle and comment.To find subtitle and whole set of comments we can add whole row and that would be distance between a subtitle to whole set of comments.

```python
distances=[]
for i in (matrix):
    distances.append(np.sum(i))
    
```

```python
distances=pd.Series(distances)
```
